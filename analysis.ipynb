```python
import json
import pandas as pd
data_file_business = open("yelp_academic_dataset_business.json",encoding="utf8")
data_business = []
for line in data_file_business:
    data_business.append(json.loads(line))
business_df = pd.DataFrame(data_business)
data_file_business.close()
```

```python
restaurants = business_df.dropna()[business_df.dropna()['categories'].str.contains('Restaurants')]
```

```python
data_file_review = open("yelp_academic_dataset_review.json",encoding="utf8")
data_review = []
for line in data_file_review:
    data_review.append(json.loads(line))
review_df = pd.DataFrame(data_review)
data_file_review.close()
```

```python
restaurants['business_id'].tolist()
```

```python
restaurants_ids = restaurants['business_id'].to_numpy()
```

```python
review_df
```

```python
comments = review_df[review_df.business_id.isin(restaurants_ids)]
```

```python
comments
```

```python
stars = comments.stars.to_numpy().astype(int)
```

```python
import matplotlib.pyplot as plt
_ = plt.hist(stars, bins = range(1,6))
plt.show()
```
